# **Understanding Data Types in Python**

```python
#Python List is More than just a List
L = list(range(20))
print("list items: ",L)
print("Type of List[0] is ",type(L[0]))
```

```python
L2 = [str(c) for c in L]
print("list items: ",L2)
print("Type of List[0] is ",type(L2[0]))
```

```python
L3 = [True,"5",3.0,3]

print("type of each item in L3",[type(i) for i in L3])
```

```python
#Creating Arrays from Python Lists
np.array([1.2,"3",3,4,5],dtype='float32')

nArr = np.array([range(i,i+3) for i in [2,4,6,4]])
print(nArr)
```

```python
#Creating Arrays from Scratch
print(np.zeros(5,dtype=np.float16))

print(np.ones((2,2),dtype=np.int16))

print(np.linspace(-1,1,10))

```

# The Basics of NumPy Arrays


```python
#NumPy Array Attributes
np.random.seed(0)

x1= np.random.randint(10,size=6)
x2 = np.random.randint(10,size=(3,5))
x3 = np.random.randint(10,size=(3,4,5))

for i in [x1,x2,x3]:
  print("\nX"+str(i.ndim))
  print(i)
  print("ndim: ",i.ndim)
  print("shape: ",i.shape)
  print("size: ",i.size)
  print("dtype: ",i.dtype)
  print("itemsize: ",i.itemsize,"bytes")
  print("nbytes:",i.nbytes,"bytes")
```

```python
#Array Indexing: Accessing Single Elements
x1
print(x1[0])
print(x1[-1])
print(x1[-3])
x3[1][1][-1] = '3399'
print(x3[1][1][-1])
```

```python
#Array Slicing: Accessing Subarrays
print(x1[:3])
print(x1[::2])
print(x1[::-1])

print("\nX2\n")
print(x2[:1,:2])

print("\nX3\n")
print(x3[:1,:2,:-1])
```

```python
#Reshaping of Arrays
x = np.arange(1,10)
print(x)
print(x.reshape(3,3))
```

```python
#Concatenation of arrays
y = np.arange(20,25)

print(np.concatenate([x,y]))
```

```python
#splitting of arrays

x1,x2 = np.split(x,[5])
print(x1,x2)
```

# Computation on NumPy Arrays: Universal Functions

```python
np.random.seed(0)

def compute_reciprocals(v):
  output = np.empty(len(v))
  for i in range(len(v)):
    output[i] = 1.0 / v[i]

  return output

v = np.random.randint(1,10,size=5)
print(v)
print(compute_reciprocals(v))
```

```python
big_array = np.random.randint(1,100,size=1000000)
%timeit compute_reciprocals(big_array)
```

```python
x = np.array([-2,1,0,-2,-1])
abs(x)
```

```python
#Trigonometric fuctions
theta = np.linspace(0,np.pi,3)
theta

print("theta = ", theta)
print("sin(theta) = ", np.sin(theta))
print("cos(theta) = ", np.cos(theta))
print("tan(theta) = ", np.tan(theta))
```

```python
#Exponents
x = [1,2,3]
print("x = ",x)
print("e^x = ", np.exp(x))
print("2^x = ", np.exp2(x))
```

# Visualization with Matplotlib


```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('classic')
```

```python
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x),'x')
plt.plot(x,np.cos(x),'s')

plt.show()
```

```python
#Two Interfaces for the Price of One

plt.figure()

plt.title("two in one")
plt.subplot(2,1,1)
plt.plot(x,np.sin(x),'_')


plt.subplot(2,1,2)
plt.plot(x,np.cos(x),'o')
```

```python
I =np.sin(x) * np.cos(x[:, np.newaxis])
plt.imshow(I)
plt.colorbar()
```
